```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flower = pd.read_csv('iris.data')
flower['flower length'] = flower['sepal length'] + flower['sepal width']
flower
```

```python
sns.barplot(x='flower length', y='species', data=flower)
print('The distinct species is iris-virginica!')
plt.title('Sepal Length-Width Combination')
```

```python
sns.catplot(
    x='petal length', 
    y='species',  
    kind='box', 
    data=flower,
)
plt.title('Distribution of Petal Length')
print('Based on the plot: ')
print('\n 1. The Iris-virginica species has no outliers when it comes to petal length')
print('\n 2. Both, Iris-versicolor & Iris-virginica both always have larger petal lengths than Iris-setosa')
```

```python
flower_correlation = flower.groupby('species').corr()
sns.heatmap(flower_correlation, annot=True, cmap='coolwarm')
plt.title('Flower Correlation Map')
print('\n The most highly correlated flower attribute are: \n Sepal Length + Sepal Width')
```
